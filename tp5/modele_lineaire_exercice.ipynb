```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data_price = pd.read_csv('price_availability.csv', sep=";")
data_price
```

```python
data_listing = pd.read_csv('listings_final.csv', sep=";")
data_listing
```

```python
data_price = data_price.groupby("listing_id").mean()
data_price
```

```python
df = pd.merge(data_price,data_listing,on=['listing_id'])
df
```

```python
x = np.array([df.person_capacity,df.bedrooms, df.bathrooms])
y = np.array(df.local_price)
```

```python
y
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x=x.reshape((999,3))
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("Coefficients beta_j :\n{}\nCoefficients INTERCEPT beta0 :\n{}".format(reg.coef_,reg.intercept_))
```

```python
yPredict = reg.predict(x_test)
```

```python
print("taille du vecteur Y_pred : {} ".format(len(yPredict)))
```

```python
print("Mean squared error: {0}\nR2 : {1}".format(mean_squared_error(y_test,yPredict),r2_score(y_test,yPredict)))
```

```python
r2_score(y_test, yPredict)
```
